```python

# Step 1: Import necessary libraries
import numpy as np # numerical python
import pandas as pd # assess dataset
import matplotlib.pyplot as plt # visualization
import seaborn as sns # better visualization and analysis
```

```python
# Step 2: Load the data
netflix_data = pd.read_csv('netflix_titles.csv')
```

```python
# Step 3: Code to make visualization better
sns.set(style="darkgrid")
```

```python
# Step 4: 'Shape' of the data: Show the total number of columns and rows - 8807 rows and 12 columns
netflix_data.shape
```

```python
# Step 5: 'head' function displays first 5 rows
netflix_data.head()
```

```python
# isnull() function - Checking for Null (NaN). 'True' is the null
netflix_data.isnull()
```

```python
# Sum of null values feild missing data by column title of the dataset
netflix_data.isnull().sum()
```

```python
# dropna(): Dropping the null value rows to avoid bias. Do it first to clean data
netflix_data_cleaned = netflix_data.dropna()
```

```python
#checking NaN after removing using dropna()
netflix_data_cleaned.isnull().sum()
```

```python
# Checking - Find out how many rows are removed and how many remaining (around 3500 rows removed)
netflix_data_cleaned.shape
```

```python
# Start the visual analysis. sns to do counting
# Univariate analysis - one variate only (here Type)
# 6 width and 4 height figure
# x - Type (movie/TVshow) and y - Count
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=netflix_data_cleaned)
plt.title('Distribution of Titles by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

**Insigt:** Movies are more prevalent than TV shows on Netflix. This indicates that Netflix has historically focused more on acquiring or producing movies compared to TV shows.<br>
<b>Recommendation</b>:Netflix could explore increasing their series content to balance the distribution.

```python
# Releases by year
plt.figure(figsize=(10,6))
sns.histplot(x='release_year', data=netflix_data_cleaned, bins=50)
# Histogram gives the frequecy or count; Bins is bars in the graph
plt.title('Number of Titles released each year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

**Insight**: Number of movies & TV Shows released per year has significant increase, especially in the last decade, indicating a growing trend in movie production and acquisition by Netflix.<br>
<b>Recommendation</b>: Netflix should continue to invest in new content to maintainand grow its subscriber base.

```python
# Bivariate analysis to seperate movies and TV show by hue='type' (seperate movies and how many TV shows by color)
plt.figure(figsize=(10,6))
sns.histplot(x='release_year', data=netflix_data_cleaned, bins=50, hue='type', multiple='stack')
# Histogram gives the frequecy or count; Bins is bars in the graph; hue-color; stack will stack
plt.title('Number of Titles released each year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

**Insight:** TV shows are present but overshadowed Still, that small orange slice is quitely growing, A hint that series content can spark long-term loyalty<br>
<b>Recommendation</b>: Fuel that momentum, Invest in standout mini-series compelling drama, and addictive reality transform casual watchers into devoted fans.

```python
# Directors produced maximum movies in Netflix. Need a specialized module form a library called 'Counter' to count the number by director.
from collections import Counter
```

```python
# Creating a function('def' to define a function in Python) to find the number of movies per director and print top '10' numbers.
def get_top_elements(column, n=10):
  all_elements = []
  for item in netflix_data_cleaned[column]:
    elements = item.split(',') #Split the string into a list
    all_elements.extend(elements) #add the list of elements to all_elements list
  element_counts = Counter(all_elements)
  top_n = element_counts.most_common(n)
  return top_n
```

```python
top_directors = get_top_elements('director', 10)
```

```python
top_directors # Textual presentation
```

```python
# To see it visually
directors, counts = zip(*top_directors)
plt.figure(figsize=(10, 6))
sns.barplot(y=directors, x=counts, hue=counts, palette='viridis')
plt.title('Top 10 Directors')
plt.xlabel('Count')
plt.ylabel('Director')
plt.show()
```

Insight: Certain directors are frequently featured.<br>
Recommendation: Netflix should continue to collaborate with popular actors and directors to attract their fanbase.

```python
# Genre (listed_in) analysis of the movies
print("\nTop 10 genres:")
top_genres = netflix_data_cleaned['listed_in'].value_counts().head(10)
print(top_genres) #Textual Analysis
```

```python
# Visual Analysis
plt.figure(figsize=(10, 6))
sns.barplot(y=top_genres.index, x=top_genres.values, hue=top_genres, palette='viridis')
plt.title('Top 10 Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

**Insight: **Dramas, International movies and stand-up comedy are the most popular genres.<br>
**Recommendtion:** Netflix should invest in these genres, as they are highly popular among viewers.

```python
# Duration analysis for movies
netflix_movies = netflix_data_cleaned[netflix_data_cleaned['type'] == 'Movie']
# Removing the string part from minutes and replacing as int (ex: 90 min)
netflix_movies['duration'] = netflix_movies['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(10, 6))
sns.histplot(netflix_movies['duration'], bins=30)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('count')
plt.show()
```

**Insight:** Most movies have a duration between 80 and 120 minutes.
<br> **Recommendation:** Netflix could consider producing more short films and mini-series to cater to viewers with less time.

```python
# Top 10 countries producing Netflix movies
top_countries_content = netflix_data_cleaned['country'].value_counts().head(10)

plt.figure(figsize=(14, 7))
sns.barplot(y=top_countries_content.index, x=top_countries_content.values, hue=top_countries_content, palette='viridis')
plt.title('Top 10 Countries with the Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

**Insight:** The USA leads in terms of content available on Netflix, followed by other countries. (This can be because Netflix is a US based company) This can help in understanding the distribution of content and guide strategic decisions to expand and diversify the content library for other regions. <br>
**Recommendation:** Diversify the content library by acquiring more titles from other countries to cater to a global audience.
